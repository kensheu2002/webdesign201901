```python
# 导入所需库
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import csv
import random

# 定义 fetch_google_news_results 函数
def fetch_google_news_results(query, keyword, driver, iteration):
    base_url = "https://news.google.com"
    search_url = f"{base_url}/search?q={query}&hl=zh-TW&gl=TW&ceid=TW:zh-Hant"

    try:
        driver.get(search_url)
        time.sleep(5)  # 等待頁面加載完全

        soup = BeautifulSoup(driver.page_source, "html.parser")
        news_articles = soup.find_all("article")

        for article in news_articles:
            headline_elem = article.find("h3", class_="ipQwMb ekueJc RD0gLb")
            if headline_elem:
                headline = headline_elem.text
            else:
                headline = "標題未找到"

            link_elem = article.find("a", class_="VDXfz")
            if link_elem:
                link = base_url + link_elem["href"]
            else:
                link = "連結未找到"

            # 檢查是否包含關鍵字，若有則輸出結果並點擊連結
            if keyword in headline:
                print("標題:", headline)
                print("文章連結:", link)

                # 提取文章來源和作者
                source_and_author = headline.split("｜", 1)[-1].strip()
                author, source = source_and_author.rsplit(" - ", 1)

                # 提取文章日期
                time_elem = article.find("time")
                if time_elem:
                    article_date = time_elem["datetime"]
                    formatted_date = time_elem.get_text(strip=True)
                    print("文章日期:", formatted_date)
                else:
                    print("文章日期未找到")

                print("文章來源:", source)
                print("文章作者:", author)
                print("=" * 50)

                # 點擊連結以增加點擊率
                driver.find_element(By.LINK_TEXT, headline).send_keys(Keys.CONTROL + Keys.RETURN)
                time.sleep(5)  # 等待頁面加載完全
                driver.switch_to.window(driver.window_handles[-1])  # 切換到新開的分頁
                print("已點擊連結:", driver.current_url)
                print("=" * 50)

                # 初始化滾動和重新整理計數
                scroll_count = 0
                refresh_count = 0

                # 隨機選擇每次捲動的秒數，1或2秒
                scroll_seconds = random.randint(1, 2)
                while scroll_count < 20:
                    if random.random() < 0.5:
                        driver.execute_script("window.scrollBy(0, -window.innerHeight);")  # 向上滚动
                    else:
                        driver.execute_script("window.scrollBy(0, window.innerHeight);")   # 向下滚动
                    time.sleep(scroll_seconds)
                    scroll_count += 1

                print(f"已進行滾動操作 {scroll_count} 次")

                # 统计往上和往下滚动次数
                scroll_up_count = 0
                scroll_down_count = 0
                for _ in range(scroll_count):
                    if random.random() < 0.5:
                        scroll_up_count += 1
                    else:
                        scroll_down_count += 1
                print(f"已進行往上滾動操作 {scroll_up_count} 次")
                print(f"已進行往下滾動操作 {scroll_down_count} 次")

                # 等待10到12秒的隨機時間
                wait_time = random.uniform(8, 10)
                time.sleep(wait_time)

                # 重新整理網頁
                refresh_count = random.randint(1, 2)  # 重新整理次數，隨機取1到2之間的數字
                for i in range(refresh_count):
                    driver.refresh()
                    sleep_time = random.randint(8, 10)  # 每次重新整理的間隔時間，隨機取8到10之間的數字
                    time.sleep(sleep_time)
                    refresh_count += 1

                print(f"已進行重新整理操作 {refresh_count} 次")

                driver.close()  # 關閉新開的分頁
                driver.switch_to.window(driver.window_handles[0])  # 切換回搜尋結果頁面

    except Exception as e:
        print("錯誤:", str(e))

if __name__ == "__main__":
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument('--headless')
    chrome_options.add_argument('--no-sandbox')
    chrome_options.add_argument('--disable-dev-shm-usage')

    # 创建 Chrome WebDriver 对象
    driver = webdriver.Chrome(options=chrome_options)

    keyword = "Kevin_malamute - 方格子"
    num_iterations = 30  # 要進行的次數

    with open("source-url-copy-8.csv", "r", encoding="utf-8-sig") as file:
        reader = list(csv.reader(file))  # 将读取器的内容转换为列表
        for iteration in range(num_iterations):
            query = random.choice([row[0] for row in reader])
            print(f"正在執行第 {iteration+1} 次迭代...")
            fetch_google_news_results(query, keyword, driver, iteration+1)

    driver.quit()
```
